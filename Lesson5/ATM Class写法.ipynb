```python
class ATM:
    def __init__(self, depositnum=10, withdrawnum=0, deposittem=0):
        self.depositnum = depositnum
        self.withdrawnum = withdrawnum
        self.deposittem = deposittem
           

class ATMsystem(ATM):
    def menu(self):
        while True:
            print('''
==========
请选择操作
1.查询余额
2.存款
3.取款
4.退出
==========
                ''')
            choice = int(input('请输入要执行功能的数字：'))

            try:
                if choice == 1:
                    self.search()
                elif choice == 2:
                    self.withdraw()
                elif choice == 3:
                    self.deposit()
                elif choice == 4:
                    print('Exit!')
                    break
                else:
                    print("输入错误！请输入1，2，3，4")
            except:
                print("输入错误!")
            

    def search(self):
        print("余额为：", self.depositnum)

    
    def withdraw(self):
        self.withdrawnum = int(input("请输入取款金额：")) 
        if self.depositnum - self.withdrawnum <0:
            print("余额不足！")
        elif self.withdrawnum <= 0:
            print("您输入的金额无效")
        else:
            self.depositnum -= self.withdrawnum 
        
    def deposit(self):
        try:
            self.deposittem = int(input("请输入存款金额："))
            if self.deposittem <= 0:
                print("您输入的金额无效")
            else:
                self.depositnum += self.deposittem
        except:
            print("只能存取整数")
    
atm = ATMsystem()
atm.menu()
```
